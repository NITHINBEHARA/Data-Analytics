```python
# prompt: handle missing values appropriately ( drop or impute them)

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Attempt to load the dataset, or create a sample DataFrame if the file is not found.
try:
  df = pd.read_csv('/content/retail_sales_dataset.csv')
except FileNotFoundError:
  print("File not found. Creating a sample DataFrame.")
  data = {'col1': [1, 2, 3, np.nan, 1,np.nan, 5,5], 'col2': ['a', 'b', 'c', 'a', 'a','b','a','a'], 'col3': [1.1, 2.2, np.nan, 1.1, 1.1,np.nan, 3.3,3.3]}
  df = pd.DataFrame(data)

# Check for missing values
print("Missing Values Before Handling:")
print(df.isnull().sum())

# Check for duplicates
print("\nDuplicate Rows Count Before Handling:")
print(df.duplicated().sum())

# Display duplicate rows if any
if df.duplicated().sum() > 0:
  print("\nDuplicate rows are:")
  print(df[df.duplicated(keep=False)])
else:
    print("\nNo duplicate rows found")

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\nDataFrame after removing duplicates:")
print(df)
print("\nDuplicate Rows Count after removing duplicates:")
print(df.duplicated().sum())

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute numerical columns with the mean
for col in numerical_cols:
  if df[col].isnull().any():
    df[col].fillna(df[col].mean(), inplace=True)

# Impute categorical columns with the mode
for col in categorical_cols:
  if df[col].isnull().any():
    df[col].fillna(df[col].mode()[0], inplace=True)

# Check for missing values after imputation
print("\nMissing Values After Handling:")
print(df.isnull().sum())
print("\nDataframe After handling missing values:")
df
```

•	Heatmap visualization
A heat map visualization is a graphical representation of data where individual values are represented by variations in color. Heat maps are commonly used to highlight patterns, trends, and concentrations within a dataset.

Key Features of Heat Maps:
Color Gradient – Uses a range of colors to indicate different values (e.g., blue for low values and red for high values).
Grid-Based or Spatial Representation – Data is often represented in a matrix/grid format, but it can also be mapped onto geographical locations.
Intensity-Based Interpretation – Darker or more intense colors usually indicate higher concentrations or values.
Common Applications:
Geographical Heat Maps – Display population density, crime rates, or disease outbreaks.
Website Analytics – Show user interactions, like click heatmaps in UX/UI analysis.
Finance & Stock Market – Represent market performance, price movements, or trading volumes.
Biological & Scientific Research – Used in genomics and brain activity studies.
Sports Analysis – Track player movements and key action areas in games.


When analyzing heat map visualizations, key correlations can be identified by observing color patterns and gradients. Here are some important insights that can be drawn from a heat map:

1. Strong Positive Correlation
If two variables show a similar color intensity in corresponding areas (e.g., both dark red in the same region), it indicates a strong positive correlation.
Example: In a stock market heat map, two sectors turning red at the same time may indicate that they rise and fall together.
2. Strong Negative Correlation
If one variable is dark (high values) while another is light (low values), they likely have a negative correlation.
Example: A heat map of crime rates vs. income levels may show high crime rates (red) in areas with lower income (blue), suggesting an inverse relationship.
3. Clusters & Patterns
Grouped areas of similar color indicate a natural clustering of data points.
